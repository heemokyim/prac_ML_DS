```python
import tensorflow as tf
```

```python
x_train = [1, 2, 3]
y_train = [1, 2, 3]

# tensorflow가 사용하는 Variable
# rank가 1인 shape으로 각 variable을 tensorflow에게 지정해준다.
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# h(x) = Wx + b
hypothesis = x_train * W + b

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - y_train))

# Gradient Descent
# minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()

# varialbe을 tf에 지정해줬다면 초기화 해줘야함
sess.run(tf.global_variables_initializer())

for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

```python
# placeholder 를 이용해 값 던져주기
X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])

W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# h(x) = Wx + b
hypothesis = X * W + b

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train], feed_dict={X: [1, 2, 3], Y: [5, 8, 11]})
    if step % 40 == 0:
        print(step, cost_val, W_val, b_val)
```

```python
# test

print(sess.run(hypothesis, feed_dict={X: [7]}))
print(sess.run(hypothesis, feed_dict={X: [-5, -7.5]}))
```

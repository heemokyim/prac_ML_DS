# matplotlib

* pyplot 객체를 사용하여 데이터를 표시
* pylot 객체에 그래프들을 쌓은 다음 show로 flush
* 최대 단점 argument를 kwargs 받음
* 그래서 matplotlib 사이트가서 찾아봐야됨

```python
import matplotlib.pyplot as plt

X = [v / 10 for v in range(100)]
Y = [v ** 3 for v in X]
plt.plot(X, Y)
plt.show()
```

* graph는 figure 객체에 생성됨
* pyplot 객체 사용시, 기본 firgure에 그래프가 그려짐

```python
import numpy as np

x_1 = [v / 10 for v in range(100)]
y_1 = [np.cos(v) for v in x_1]

x_2 = [v / 10 for v in range(100)]
y_2 = [np.sin(v) for v in x_2]

line1, = plt.plot(x_1, y_1, label="cos")
line2, = plt.plot(x_2, y_2, label="sin")
plt.legend(handles=[line1, line2], loc="lower right") # legend
plt.title('test title') # title
plt.show()
```

```python
X = [v / 10 for v in range(100)]
Y = [10 - v for v in X]

plt.plot(X, Y)
plt.plot(x_1, y_1)
plt.plot(y_2, x_2)
plt.show() # memory flush => draw
```

```python
fig = plt.figure() # figure return
fig.set_size_inches(10, 5) # 크기 지정
ax_1_1 = fig.add_subplot(1, 2, 1) # (1,1) in (2,2)
ax_1_2 = fig.add_subplot(1, 2, 2) # (1,2) in (2,2)


ax_1_1.plot(x_1, y_1, c="b") # #eeefff 이런 식으로 컬러지정 가능
ax_1_2.plot(x_2, y_2, c="g")
plt.show()

fig = plt.figure() # figure return
fig.set_size_inches(10, 5) # 크기 지정
ax_2_1 = fig.add_subplot(1, 2, 1)
ax_2_2 = fig.add_subplot(1, 2, 2)

ax_2_1.plot(x_1, y_1, c="b", linestyle='dashed')
ax_2_1.plot(x_2, y_2, c="g", linestyle='dotted')
plt.title('clock')
ax_2_2.plot(x_1, y_1, c="b")
ax_2_2.plot(y_1, x_1, c="g")
ax_2_2.plot(Y, X, c='r')
plt.show()

```
